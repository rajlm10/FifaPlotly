```python
import pandas as pd
import numpy as np
```

```python
df_dict={'2015':pd.read_csv("players_15.csv"),
         '2016':pd.read_csv("players_16.csv"),
         '2017':pd.read_csv("players_17.csv"),
         '2018':pd.read_csv("players_18.csv"),
         '2019':pd.read_csv("players_19.csv"),
         '2020':pd.read_csv("players_20.csv"),
         '2021':pd.read_csv("players_21.csv"),
}
```

```python
df_dict['2015'].head()
```

```python
combined=pd.DataFrame()
#combined.append(temp)
```

```python
for year in df_dict.keys():
  temp=df_dict[year]
  temp['year']=year
  combined=combined.append(temp,ignore_index=True)
```

```python
from collections import Counter
```

```python
fifa_appearances=Counter(combined['sofifa_id'])
```

```python
ids_list=[]
for id,count in fifa_appearances.items():
  if(count==7):
    ids_list.append(id)
  if(len(ids_list)==20):
    break
```

```python
final_df=combined[combined['sofifa_id'].isin(ids_list)]
```

```python
positions_dict={'Attacker':['CF','RF','LF','ST','LW','RW'],
                'Midfielder':['LM','RM','CAM','CM','CDM'],
                'Defender':['RB','LB','CB','LWB','RWB'],
                'Goalkeeper':['GK']
                }
```

```python
def return_position(pos):
  if pos in positions_dict['Attacker']:
    return 'Attacker'
  elif pos in positions_dict['Midfielder']:
    return 'Midfielder'
  elif pos in positions_dict['Defender']:
    return 'Defender'
  elif pos in positions_dict['Goalkeeper']:
    return 'Goalkeeper'
```

```python
final_df['player_positions']=final_df['player_positions'].apply(lambda x: return_position(x.split(',')[0]))
```

```python
final_df=final_df[['sofifa_id','player_url','short_name','age','overall','value_eur','player_positions','year']]
```

```python
final_df=final_df.rename(columns={'player_positions':'player_position'})
```

```python
final_df.to_csv("combined_fifa.csv")
```
